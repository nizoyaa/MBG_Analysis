```python
import pandas as pd
import requests

#Unduh kamus leksikon positif dan negatif dari github
positive_url = "https://raw.githubusercontent.com/fajri91/InSet/master/positive.tsv"
negative_url = "https://raw.githubusercontent.com/fajri91/InSet/master/negative.tsv"

positive_lexicon = set(pd.read_csv(positive_url, sep='\t', header=None)[0])
negative_lexicon = set(pd.read_csv(negative_url, sep='\t', header=None)[0])

def determine_sentiment(text):
  if isinstance(text, str):
    positive_count = sum(1 for word in text.split() if word in positive_lexicon)
    negative_count = sum(1 for word in text.split() if word in negative_lexicon)
    sentiment_score = positive_count - negative_count
    sentiment = "Negatif" if sentiment_score <= 0 else "Positif"
    return sentiment_score, sentiment
  return 0, "Negatif"

data[['Score', 'Sentiment']] = data['stemming_data'].apply(lambda x: pd.Series(determine_sentiment(x)))
data.head(5)
```

**LDA Topic Modeling**

```python
#Tahap 1  Import Library

import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Tahap 2  Persiapkan Data

data = pd.DataFrame(data[['stemming_data']]).dropna()
data.head(5)
```

```python
#Tahap 3  Vectorization (Mengubah Teks ke Angka)

from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(
    max_df=0.9,    # abaikan kata yang muncul di >90% dokumen
    min_df=2,      # abaikan kata yang muncul di <2 dokumen
    stop_words=None  # tidak menggunakan builtin stopword
)

X = vectorizer.fit_transform(data['stemming_data'])
print("Shape matrix:", X.shape)
```

```python
#Tahap 4  Membangun Model LDA

n_topics = 5  # kamu bisa ubah sesuai kebutuhan (misal 58 topik)

lda_model = LatentDirichletAllocation(
    n_components=n_topics,  # jumlah topik
    random_state=42,
    learning_method='batch'
)
lda_model.fit(X)
```

```python
#Tahap 5  Menampilkan Kata Kunci Tiap Topik

def display_topics(model, feature_names, n_top_words):
    for topic_idx, topic in enumerate(model.components_):
        print(f"Topik {topic_idx}:")
        print(" | ".join([feature_names[i] for i in topic.argsort()[:-n_top_words - 1:-1]]))
        print()

n_top_words = 10
tf_feature_names = vectorizer.get_feature_names_out()
display_topics(lda_model, tf_feature_names, n_top_words)
```

```python
#Tahap 6  Memberi Label Topik ke Setiap Berita

topic_values = lda_model.transform(X)
data['Dominant_Topic'] = topic_values.argmax(axis=1)
data.head()
```

```python
#Tahap 7  Visualisasi Distribusi Topik

plt.figure(figsize=(8,5))
sns.countplot(x='Dominant_Topic', data=data, palette='coolwarm')
plt.title('Distribusi Jumlah Berita per Topik (LDA)')
plt.xlabel('Topik')
plt.ylabel('Jumlah Berita')
plt.show()
```

```python
# Ambil 10 kata teratas dari setiap topik
feature_names = vectorizer.get_feature_names_out()
n_top_words = 10

topic_keywords = []
for topic_idx, topic in enumerate(lda_model.components_):
    top_words = [feature_names[i] for i in topic.argsort()[:-n_top_words - 1:-1]]
    topic_keywords.append({
        'Topic': topic_idx,
        'Keywords': ', '.join(top_words)
    })

# Buat DataFrame berisi topik dan kata-katanya
df_topics = pd.DataFrame(topic_keywords)
df_topics
```

```python
def auto_label(keywords):
    keywords_lower = keywords.lower()
    if any(word in keywords_lower for word in ['racun', 'korban', 'sakit', 'alami']):
        return 'Kasus Keracunan'
    elif any(word in keywords_lower for word in ['gizi', 'sppg', 'siswa', 'makan']):
        return 'Program Gizi Sekolah'
    elif any(word in keywords_lower for word in ['program', 'gratis', 'presiden', 'perintah']):
        return 'Kebijakan Pemerintah'
    elif any(word in keywords_lower for word in ['rp', 'anggar', 'triliun', 'ekonomi']):
        return 'Anggaran dan Ekonomi'
    elif any(word in keywords_lower for word in ['ikan', 'susu', 'bakteri', 'kandung']):
        return 'Kandungan dan Kesehatan Makanan'
    else:
        return 'Lainnya'

# Terapkan fungsi auto-label
df_topics['Auto_Label'] = df_topics['Keywords'].apply(auto_label)
df_topics
```

```python
# Ambil distribusi topik tiap dokumen
topic_distribution = lda_model.transform(X)

# Tentukan topik dominan untuk setiap dokumen
dominant_topics = topic_distribution.argmax(axis=1)

# Gabungkan dengan data asli
data_labeled = data.copy()
data_labeled['Dominant_Topic'] = dominant_topics

# Tambahkan label otomatis
data_labeled = data_labeled.merge(df_topics[['Topic', 'Auto_Label']],
                                  left_on='Dominant_Topic', right_on='Topic', how='left')

# Tampilkan hasil akhir
data_labeled[['stemming_data', 'Dominant_Topic', 'Auto_Label']].head(10)
```

```python
from datetime import datetime, timedelta
import re

# Kamus bulan Indonesia
bulan_ind = {'Jan':1, 'Feb':2, 'Mar':3, 'Apr':4, 'Mei':5, 'Jun':6,
             'Jul':7, 'Agu':8, 'Sep':9, 'Okt':10, 'Nov':11, 'Des':12}

def parse_waktu_id(text):
    text = str(text).strip()

    # Relative time
    if "menit yang lalu" in text:
        menit = int(re.search(r'(\d+)', text).group(1))
        return datetime.now() - timedelta(minutes=menit)
    elif "jam yang lalu" in text:
        jam = int(re.search(r'(\d+)', text).group(1))
        return datetime.now() - timedelta(hours=jam)

    # Absolute time format bahasa Indonesia
    match = re.search(r'\w+,\s*(\d{2}) (\w+) (\d{4}) (\d{2}):(\d{2})', text)
    if match:
        hari, hari_bulan, tahun, jam, menit = match.groups()
        bulan = bulan_ind.get(hari_bulan, 0)
        if bulan > 0:
            return datetime(int(tahun), int(bulan), int(hari), int(jam), int(menit))

    return pd.NaT

# Terapkan parsing
pre['waktu_terbit_fix'] = pre['waktu_terbit'].apply(parse_waktu_id)
pre['tanggal'] = pre['waktu_terbit_fix'].dt.date

print(pre[['waktu_terbit', 'waktu_terbit_fix', 'tanggal']].head(20))
print("Jumlah baris valid:", pre['tanggal'].notna().sum())
```

```python
# Gabungkan kembali dengan LDA
gabung = pd.merge(lda, pre[['stemming_data', 'tanggal']], on='stemming_data', how='left')

# Hitung jumlah posting per topik per tanggal
trend = gabung.groupby(['tanggal', 'Topic_Label']).size().reset_index(name='jumlah')

# Visualisasi tren per topik
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(14,7))
sns.lineplot(data=trend, x='tanggal', y='jumlah', hue='Topic_Label', marker='o')
plt.title("Trend Topik MBG per Tanggal")
plt.xlabel("Tanggal")
plt.ylabel("Jumlah Posting")
plt.xticks(rotation=45)
plt.legend(title='Topik')
plt.tight_layout()
plt.show()
```

```python
pivot = trend.pivot(index='Topic_Label', columns='tanggal', values='jumlah').fillna(0)
plt.figure(figsize=(14,6))
sns.heatmap(pivot, cmap='YlOrRd', linewidths=.5, annot=True, fmt='.0f')
plt.title("Heatmap Jumlah Posting per Topik per Tanggal")
plt.xlabel("Tanggal")
plt.ylabel("Topik")
plt.show()
```
